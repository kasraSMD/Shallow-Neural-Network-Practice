````python
train_data = pd.read_csv(r"data\diabetes_train.csv") 
train_data.head()
````

````python
test_data = pd.read_csv(r"data\diabetes_test.csv") 
test_data.head()
````

````python
train_data_outcome = train_data["Outcome"] # TODO
train_data = train_data.drop(columns=["Outcome"]) # TODO: drop Outcome

train_data.head()
````

### Z-Score Normalization
$$ Z = \frac{x_i - \bar{x}}{\sigma} $$


````python
for column in train_data.columns:
  mean = train_data[column].mean() # TODO (Mean of train_data[column])
  std = train_data[column].std() # TODO (Standard Deviation of train_data[column])
  train_data[column]=(train_data[column]- mean)/std
  test_data[column]=(test_data[column] - mean)/std
  
    
train_data
````

````python
test_data
````

### add column 1 as bias

````python
train_num_rows = train_data.shape[0] 
train_bias = np.ones(train_num_rows).reshape(-1, 1) # TODO
train_data = pd.DataFrame(train_bias, columns=['Bias']).join(train_data) # TODO: add Bias to train_data

test_num_rows = test_data.shape[0] 
test_bias = np.ones(test_num_rows).reshape(-1, 1) # TODO
test_data = pd.DataFrame(test_bias, columns=['Bias']).join(test_data) # TODO: add Bias to test_data

train_data
````

### train_test_split

````python
from sklearn.model_selection import train_test_split

X_train, X_validation, y_train, y_validation = train_test_split(train_data.to_numpy(), train_data_outcome.to_numpy(), 
                                                                test_size=0.2) 

X_train = X_train.T # TODO
X_validation = X_validation.T # TODO
y_train = y_train.T # TODO
y_validation = y_validation.T # TODO
test_data_numpy = test_data.T.to_numpy() # TODO
````

### Create Model

````python
class Model:

    def __init__(self):

        self.w1 = np.random.randn(1000, 9) * 0.01 # TODO: random matrix with size (1000, number_of_input_feautres)
        self.w2 = np.random.randn(1,1000) * 0.01 # TODO: random matrix with size (1, 1000)

    def predict(self, inputs):
        x = inputs

        Z_1 = self.w1 @ x # TODO
        
        A_1 = np.maximum(0, Z_1) # TODO
        
        Z_2 = self.w2 @ A_1 # TODO
        
        A_2 = 1 / (1 + np.exp(-Z_2)) # TODO
        
        return A_1, A_2

    def update_weights_for_one_epoch(self, inputs, outputs, learning_rate):
        x, y_true = inputs, outputs
        A_1, A_2 = self.predict(inputs)

        n = inputs.shape[1] # TODO (n = number of samples)
        
        shared_coefficient = None # TODO
        relu_gradient = np.where(A_1 > 0, 1, 0)
        

        self.w1 = self.w1 + ((((2 * learning_rate / n) * (outputs - A_2) * A_2 * (1 - A_2)).T @ self.w2).T * relu_gradient) @ x.T
        self.w2 = self.w2 + ( (2 * learning_rate / n) * (outputs - A_2) * A_2 * (1 - A_2)) @ A_1.T # TODO

    def fit(self, inputs, outputs, learning_rate, epochs=64):
        for i in range(epochs):
            self.update_weights_for_one_epoch(inputs, outputs, learning_rate)
````

### Train the Model

````python
def evaluation(model, inputs, outputs):
  _, A_2 = model.predict(inputs)
  prediction = (A_2 > 0.5)
  return np.mean(prediction == outputs) * 100
````

````python
model = Model()
model.fit(X_train, y_train, learning_rate = 0.01, epochs = 64) # TODO

# Model evaluation 
print(f"you model accuracy on given set: {round(evaluation(model, X_validation, y_validation), 2)}%") # TODO
````
